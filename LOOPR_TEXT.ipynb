```python
with open('/content/TEXT_FILE_OF_10MB.txt') as f:
    lines = f.readlines()
    #print(lines)
var2={}                        #HERE WE ARE EXTRACTING THE UNIQUE VALUES 
for i in lines:
  if len(i)>0:
    x = i.split()

    for j in x:
      if j in var2:
        var2[j]+=1
      else:
        var2[j]=1

a=0
for i in var2:
  if var2[i]>a:
    a=var2[i]
    most_freq=i              #FINDING WHICH IS MOST FERQUENTLY REPEATED WORD


with open('/content/TEXT_FILE_OF_10MB.txt', 'r') as file :
  filedata = file.read()

# REPLACING THE MOST FREQUENT WORD WITH LOOPR
filedata = filedata.replace(most_freq, 'loopr')

#UODATING THE FILE
with open('/content/TEXT_FILE_OF_10MB.txt', 'w') as file:
  file.write(filedata)          
```
